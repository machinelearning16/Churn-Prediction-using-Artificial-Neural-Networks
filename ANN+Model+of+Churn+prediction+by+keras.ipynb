# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:/Users/gagan/Downloads/Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
```

```python
X
```

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initializing Neural Network
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
```

```python
# Adding the output layer(for more than two categories in dependent variable use activation=SoftMax function.)
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network(for more than two categories in dependent variable use loss=categorical_crossentropy)
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
```

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
